```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

```

Reading CSV

```python
df = pd.read_csv('car data.csv')
```

Going Through data

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Data Cleaning

```python
input = df.drop(['Car_Name','Selling_type','Owner'],axis='columns')

target = df.Selling_Price
target
```

Changing string into numbers #Encoding

```python
from sklearn.preprocessing import LabelEncoder
Numerics = LabelEncoder()
```

#New encoded columns

```python
input['Fuel_Type_n']=Numerics.fit_transform(input['Fuel_Type'])
input['Transmission_n']=Numerics.fit_transform(input['Transmission'])
input
```

```python
input_1 = input.drop(['Fuel_Type','Transmission','Selling_Price'],axis='columns')
input_1
```

```python
#Linear regression
model=linear_model.LinearRegression()
```

```python
#Training
model.fit(input_1,target)
```

```python
#prediction of car price
prediction_1 = model.predict([[2011,4.15,5200,2,1]])
prediction_2 = model.predict([[2014,5.59,27000,2,1]])
print(prediction_1)
print(prediction_2)
```
